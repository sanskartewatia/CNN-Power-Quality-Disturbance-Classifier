```python
#PLAIN 
t=np.arange(0,0.2,0.001)
#t=np.linspace(0, 1, 1000)
f=50
main0=[]
dstb0=[]
for i in range(300):
    theta=(round(random.uniform(-3.13,3.13), 3))
    magn=round(random.uniform(-5,5),2)
    y0=np.sin(2*np.pi*f*t+theta)
    main0.append(y0)
    dstb0.append(0)


for i in range(len(main0)):
    for j in range((200)):
        main0[i][j]=main0[i][j]+np.random.normal(0,0.02)
        main0[i][j]*=220
dict0={'Signal':main0,'Class':dstb0,'Sag':'0','Swell':'0','Intrpn':'0','Harmonic':'0','Flicker':'0','Transient':'0'}
df=pd.DataFrame(dict0)
display(df)
```

```python
main1=[]
dstb1=[]
#SAG Data
t=np.arange(0,0.2,0.001)
f=50
alp=[]
ta=[0.04,0.06,0.08]
tb=[0.12, 0.16, 0.14]
alp=[]
for i in range(20):
    alpha=np.random.normal(0.6, 0.067)
    alp.append(alpha)

for i in range(len(alp)):
    for j in range(len(ta)):
        for k in range (len(tb)):
            phi=(round(random.uniform(-3.13,3.13), 3))
            y1= (1-alp[i]*(np.heaviside((t-ta[j]), 0.5)-np.heaviside((t-tb[k]),0.5))) * np.sin(2*np.pi*f*t+phi)
            main1.append(y1)
            dstb1.append(1)


for i in range(len(main1)):
    for j in range((200)):
        main1[i][j]=main1[i][j]+np.random.normal(0,0.02)
        main1[i][j]*=220
        
dict1={'Signal':main1,'Class':dstb1,'Sag':1,'Swell':0,'Intrpn':0,'Harmonic':0,'Flicker':0,'Transient':0}
df1=pd.DataFrame(dict1)
#display(df1)
df=df.append(df1, ignore_index=True)
display(df)
```
